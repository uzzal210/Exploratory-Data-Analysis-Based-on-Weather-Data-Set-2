# Exploratory Data Analysis (EDA) is the discovery of unseen patterns within the whole data


# The primary purpose of exploratory data analysis is:

Investigating datasets and various variables.
Checking for null values ​​and other unexpected values.
Learn about descriptive statistics of data sets.
Knowing about the relationship between different variables.
Extracting various hidden information.
Making decisions about machine learning models.


The feature of our data set

    1.YEAR- Observation year 
    2.Month- Month
    3.MaxTemp- Mean Maximum temprature per month (Unit-celsius)
    4.MinTemp- Mean Minimum temprature per month (Unit-celsius)
    4.RelativeHumidity- Mean RelativeHumidity per month (Unit-%)
    5.Rainfall- Total Rainfall per month (Unit-mm)

# Overall Time Seris plot using matplotlib


```python
import matplotlib.pyplot as plt

df.plot(figsize=(20,10))
plt.title("Time series plot of Dhaka weather")
plt.show()
     
```

# MaxTemp Time Series plot using sns


```python
plt.figure(figsize=(16, 6))
plt.title('Time series plot of MaxTemp')
ax = sns.lineplot(x="YEAR", y="MaxTemp",err_style="bars",label="MaxTemp", ci=95,data=df)
     
```

# MinTemp Time Series plot using sns


```python
plt.figure(figsize=(16, 6))
plt.title('Time series plot of MinTemp')
ax = sns.lineplot(x="YEAR", y="MinTemp",color="blue", ci=90,label="MinTemp",data=df)
```

# Relative HumidityTS plot using sns


```python
plt.figure(figsize=(16, 6))
plt.title('Time series plot of Relative Humidity')
ax = sns.lineplot(x="YEAR", y="RelativeHumidity",label="Relative Humidity",ci=80,data=df)
```
